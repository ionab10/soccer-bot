```python
import numpy as np
from matplotlib import pyplot as plt
import cv2


class FieldView:

    def __init__(self, img):
        self.orig_img = img
        self.img = img
        
    def resize(self, height):
        self.img = imutils.resize(self.img, height=height)
        
    def increase_contrast(self):
        lab = cv2.cvtColor(self.img, cv2.COLOR_BGR2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
        cl = clahe.apply(l)
        limg = cv2.merge((cl,a,b))
        self.img = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)
        
    def get_mask(self, thresh=200):
        gray = cv2.cvtColor(self.img, cv2.COLOR_BGR2GRAY)
        blurred = cv2.GaussianBlur(gray, (5, 5), 0)
        return cv2.threshold(blurred, thresh, 255, cv2.THRESH_BINARY)[1]
        
    def get_contours(self):
        cnts = cv2.findContours(self.get_mask().copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        return imutils.grab_contours(cnts)
    
    def extract_lines(self, mask_thresh=200, min_len=75, thresh=100):
        edges = cv2.Canny(self.get_mask(thresh=mask_thresh),50,150,apertureSize = 3)
        lines = cv2.HoughLinesP(image=edges,rho=1,theta=np.pi/180, 
                                threshold=thresh,lines=np.array([]), 
                                minLineLength=min_len, 
                                maxLineGap=80)
        return edges, lines

```

```python
for fn in glob('./samples/*'):
    print(fn)

    fig, axs = plt.subplots(1, 4)

    img = cv2.imread(fn, cv2.IMREAD_COLOR)
    img = FieldView(img) 
    img.resize(210)

    #img.increase_contrast()
    axs[0].imshow(img.img)
    axs[1].imshow(img.get_mask(thresh=200))

    edges, lines = img.extract_lines(thresh=85, min_len=20)
    axs[2].imshow(edges)
    annotated = img.img.copy()
    if lines is not None:
        for line in lines:
            cv2.line(annotated, (line[0][0], line[0][1]), (line[0][2], line[0][3]), (255, 0, 0), 3, cv2.LINE_AA)
    axs[3].imshow(annotated)

    plt.show()
```
